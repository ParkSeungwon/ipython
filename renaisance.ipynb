```python
import matplotlib.pyplot as plt
```

```python
x = [0, 10, 20, 30, 40]
y = [100, 50, 80, 40, 70]
plt.plot(x, y)
ax = plt.gca()
ax.text(0, 100, 'Buddha')
ax.text(20, 80, 'Nagarjuna')
ax.text(40, 70, 'Zen')
ax.text(20, 60, '3 marks of existence')
ax.axis('off')
x1 = [0, 50]
x2 = [65, 65]
plt.plot(x1, x2, 'red')
plt.savefig('3mark.png', dpi=300)
plt.show()
```
