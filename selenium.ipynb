```python
def get_url(url):
    drv = webdriver.PhantomJS()
    drv.get(url)
    doc = drv.page_source.encode('utf-8')
    drv.service.process.send_signal(signal.SIGTERM)
    return doc
```

```python
def get_date():
    today = datetime.date.today()
    return today
```
