```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = np.arange(10,100000,50);
w = 2 * np.pi * f;
```

```python
h1 = np.abs(1 / (1+(1j * w)/4000));
h2 = np.abs(1 / (1 + 2000/(1j * w)));
h3 = np.abs(1 / (1j * w / 3000+1000/(1j * w)+1));
```

```python
plt.xlim(10, 10e+6)
plt.ylim(0, 1.1)
plt.semilogx(w, h1,w, h2, w, h3)
```

```python
plt.show()
```
